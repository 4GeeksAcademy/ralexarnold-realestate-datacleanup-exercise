#### Exercise 10. Are the averages of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# TODO
pop1 = 'Valdemorillo'
pop2 = 'Galapagar'

df['pps'] = df['price']/df['surface']

condition1 = df['level5']==pop1
condition2 = df['level5']==pop2

df.loc[condition1, ['price', 'surface', 'pps']]

pop1_mean_pps = df.loc[condition1, 'pps'].mean()
pop2_mean_pps = df.loc[condition2, 'pps'].mean()

pop1_mean_pps, pop2_mean_pps
```
